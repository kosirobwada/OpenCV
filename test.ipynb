```python
import pandas as pd
import matplotlib.pyplot as plt

# 相関行列を作成
correlation_matrix = df.corr()

# 'Attrition' 列に対する相関係数の絶対値を取得
attrition_correlation = correlation_matrix['Attrition'].abs()

# 'BusinessTravel' 列を除外
attrition_correlation = attrition_correlation.drop(['BusinessTravel'])
# 'Attrition' 列を除外
attrition_correlation = attrition_correlation.drop(['Attrition'])

# 特徴量の重要度をプロット
plt.figure(figsize=(12, 6))
attrition_correlation.plot(kind='bar')
plt.xlabel('Feature')
plt.ylabel('Absolute Correlation with Attrition')
plt.title('Feature Importance')
plt.xticks(rotation=90)
plt.show()
```

```python
# "attrition"カラムの0と1の数を数える
attrition_counts = df['Attrition'].value_counts()

# それぞれの割合を計算する
attrition_ratio = attrition_counts / attrition_counts.sum()

# 円グラフを描画する
plt.figure(figsize=(5, 5))
plt.pie(attrition_ratio, labels=attrition_ratio.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # アスペクト比を保つ
plt.title('Attrition ratio')
plt.show()
```

```python
# "attrition"カラムの値が0の行の数をカウントする
attrition_0_sum = df['Attrition'].value_counts().get(0, 0)

print("attritionが0の人の合計:", attrition_0_sum)
```

```python
# "attrition"カラムの値を文字列に変換して、0を"No"、1を"Yes"に変換する
df['attrition_str'] = df['Attrition'].map({0: 'No', 1: 'Yes'})

# "attrition_str"カラムの値ごとにグループ化して合計を計算する
attrition_counts = df['attrition_str'].value_counts()

# グラフを描画する
plt.bar(attrition_counts.index, attrition_counts.values, color=['blue', 'orange'])
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.title('Attrition Count (No vs Yes)')
plt.show()
```

```python
# "MonthlyAchievement"カラムの平均を計算する
average_achievement = df['MonthlyAchievement'].mean()

print("MonthlyAchievementの平均:", average_achievement)
```

```python
import pandas as pd
import lightgbm as lgb
import matplotlib.pyplot as plt

# 特徴量とターゲットの分割
X = df.drop(columns=['Attrition'])
y = df['Attrition']

# "attrition_str"カラムをOne-Hotエンコーディングして数値データに変換
X = pd.get_dummies(X)

# LightGBMのモデルを定義
model = lgb.LGBMClassifier()

# モデルの学習
model.fit(X, y)

# 特徴量の重要度を取得
feature_importances = model.feature_importances_

# 特徴量名と重要度を対応させるためにデータフレームに変換
feature_importances_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# 重要度で降順にソート
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

# グラフで重要度を表示
plt.figure(figsize=(6, 4))
plt.bar(feature_importances_df['Feature'], feature_importances_df['Importance'])
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importance by LightGBM')
plt.xticks(rotation=90)  # X軸のラベルを傾けて表示
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# 特徴量とターゲットの分割
X = df.drop(columns=['Attrition'])
y = df['Attrition']

# ランダムフォレストのモデルを定義
model = RandomForestClassifier()

# モデルの学習
model.fit(X, y)

# 特徴量の重要度を取得
feature_importances = model.feature_importances_

# 特徴量名と重要度を対応させるためにデータフレームに変換
feature_importances_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# 重要度で降順にソート
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

# グラフで重要度を表示
plt.figure(figsize=(6, 4))
plt.bar(feature_importances_df['Feature'], feature_importances_df['Importance'])
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importance by Random Forest')
plt.xticks(rotation=90)  # X軸のラベルを傾けて表示
plt.show()
```

```python
# Ageのヒストグラムを表示
plt.figure(figsize=(8, 5))
plt.hist(df[df['Attrition'] == 0]['Age'], bins=20, alpha=0.5, label='Attrition=No')
plt.hist(df[df['Attrition'] == 1]['Age'], bins=20, alpha=0.5, label='Attrition=Yes')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Attrition by Age')
plt.legend()
plt.show()

# MonthIncomeのヒストグラムを表示
plt.figure(figsize=(8, 5))
plt.hist(df[df['Attrition'] == 0]['MonthlyIncome'], bins=20, alpha=0.5, label='Attrition=No')
plt.hist(df[df['Attrition'] == 1]['MonthlyIncome'], bins=20, alpha=0.5, label='Attrition=Yes')
plt.xlabel('MonthlyIncome')
plt.ylabel('Count')
plt.title('Attrition by MonthlyIncome')
plt.legend()
plt.show()

# Overtimeの割合を円グラフで表示
attrition_overtime_count = df.groupby(['Attrition', 'OverTime']).size().reset_index(name='Count')
plt.figure(figsize=(5, 5))
plt.pie(attrition_overtime_count['Count'], labels=attrition_overtime_count['OverTime'], autopct='%1.1f%%', startangle=90)
plt.title('Attrition Overtime Percentage')
plt.show()
```

```python
# MonthlyIncomeが2500以下のデータを抽出
data_filtered = df[df['MonthlyIncome'] <= 2500]

# MonthlyIncomeに500を加える
data_filtered['MonthlyIncome'] += 500

# 特徴量とターゲットの分割
X = data_filtered.drop(columns=['Attrition'])
y = data_filtered['Attrition']

# ランダムフォレストのモデルを定義
model = RandomForestClassifier()

# モデルの学習
model.fit(X, y)

# 特徴量の重要度を取得
feature_importances = model.feature_importances_

# 特徴量名と重要度を対応させるためにデータフレームに変換
feature_importances_df = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importances})

# 重要度で降順にソート
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

# グラフで重要度を表示
plt.figure(figsize=(6, 4))
plt.bar(feature_importances_df['Feature'], feature_importances_df['Importance'])
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Feature Importance by Random Forest (MonthlyIncome + 500)')
plt.xticks(rotation=90)  # X軸のラベルを傾けて表示
plt.show()

# Attritionの総数を表示
attrition_count = data_filtered['Attrition'].count()
print("Attritionの総数:", attrition_count)
```
